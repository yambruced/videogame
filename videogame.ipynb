# Top 10 Systems with the most games sold

```python
plt.figure(figsize=(8,6))
plt.title("Gaming Systems with the most game sales worldwide")
plt.xlabel('Global Sales')
V.groupby('Platform')['Global_Sales'].mean().sort_values(ascending=False).head(20).plot.barh(alpha=.7)
```

# Top 10 Selling Videogames of all time

```python
plt.figure(figsize=(8,6))
plt.title("Top Selling Games of all time")
plt.xlabel('Sales Worldwide')
V.groupby('basename')['Global_Sales'].mean().sort_values(ascending=False).head(10).plot.barh(alpha=.7)
```

# Top 10 best selling publishers 

```python
plt.figure(figsize=(8,6))
plt.title("Top 10 Best Selling Videogame Publishers")
plt.xlabel('Sales')
V.groupby('Publisher')['Global_Sales'].mean().sort_values(ascending=False).head(10).plot.barh(alpha=.7)
```

# Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.title("Total Games Shipped\nfrom 1982 to 2020")
plt.xlabel('Year')
plt.ylabel('Total Shipped in Millions')
plt.scatter(x=V['Year'],y=V['Total_Shipped'],alpha=.7)
```

```python
plt.figure(figsize=(8,6))
plt.title("The Global Sales of videogames\nfrom 1970 to 2020")
plt.xlabel('Year')
plt.ylabel('Global Sales in Millions')
plt.scatter(x=V['Year'],y=V['Global_Sales'],alpha=.7)
```

```python
plt.style.use('ggplot')
data = pd.DataFrame({"year" : [2015,2016,2017,2018,2019],
                     "count" :[0.386486,0.344286,0.293324,0.248745,0.078065] })
ax = data.plot(x="year",y="count",linestyle='dashed',marker='o',alpha=.6)
plt.ylabel('Global Sales')
plt.xlabel('Year')
ax.locator_params(integer=True)
plt.title("Average Videogame Global\nSales in the last five years")
plt.tight_layout()
plt.show()
```
